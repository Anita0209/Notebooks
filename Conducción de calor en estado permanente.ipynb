## Conducción de calor en estado estacionario unidimensional a través de Placa plano
![imagen de placa plana](https://github.com/Anita0209/imagenes/blob/main/Placa%20plana.png?raw=true)
<div style="text-align: justify">Considere el sistema que se muestra arriba. La parte superior, inferior, frontal y
La parte posterior del cubo está aislada, de modo que se pueda conducir el calor
a través del cubo solo en la dirección X. En este caso especial, calor
El flujo es unidimensional. Si los lados no estuvieran aislados, el flujo de calor podría
ser bidimensionales o tridimensionales.
    
$$ T = T_{1}  cuando x = 0 $$
    
$$ T = T_{3}  cuando x = L$$
    
$$\frac{d}{dx}(\frac{dT}{\delta x})=0$$

Integrando obtenemos 
    $$\frac{dT}{dx} = C_{1}$$
Integrando de nuevo obtenemos
    $$T = C_{1}x+C_{2}$$
 En la siguiente celda verás el código en python para graficar esta ecuación,  al correrlo obtendrás la ecuación.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

output_notebook()


# Definir los coeficientes C1 y C2 Los puedes cambiar y bolver a correr
# para ver el camnio de la grafica
C1 = 80
C2 = 52

# Generar algunos valores de x
x = np.linspace(0, 10, 100)

# Calcular los valores correspondientes de T usando la ecuación
T = C1 * x + C2

# Crear una figura
fig = figure(title='Gráfico de T = C1x + C2', x_axis_label='Velocidad', y_axis_label='Temperatura')

# Agregar una línea para los datos de T
fig.line(x, T, legend_label='T', line_width=2)

# Mostrar la figura
show(fig)
```

# Conducción de calor en estado estacionario unidimensional a través de Superficie cilíndrico

![imagen de placa plana](https://github.com/Anita0209/imagenes/blob/main/superficie%20cilindrica.png?raw=true)
$$T = T_{1} cuando r = a$$
$$T = T_{2} cuando r = b$$

$$\frac{d}{dr}(r\frac{dT}{dr}) = 0$$

integrando obtenemos

$$r\frac{dT}{dr} = C_{1}$$

$$\frac{dT}{dr} = \frac{C_{1}}{r}$$

integrando de nuevo obtenemos
$$T = C_{1} lnr+C_{2}$$


Usando la condición de frontera en r = a, b y y b>a se obtiene

$$T_{1} = C_{1}lna+C_{1}$$
$$T_{2} = C_{2}lnb+C_{2}$$

Ahora en la siguiente esta el código de python para ejemplificar con una gráfica las ecuaciónes anteriores.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

output_notebook()

#Asignando valor a las variable , puedes cambiar a y b siempre y cuando b>a
a = 1
b = 10
c1 = 2
c2 = 5

# Crear una lista de valores de x que van desde a hasta b
x_values = np.linspace(a, b, num=100)

# Calcular los valores correspondientes de T1 y T2
T1_values = -c1/x_values + c2
T2_values = c1 * np.log(x_values) + c2

# Crear una figura
fig = figure(title='T1 y T2', x_axis_label='x', y_axis_label='T')

# Agregar las líneas de T1 y T2 a la figura
fig.line(x_values, T1_values, legend_label='T1', line_width=2)
fig.line(x_values, T2_values, legend_label='T2', line_width=2, line_color='red')

show(fig)
```

# Conducción de calor en estado estacionario unidimensional a través de Superficie esférico

![imagen de placa plana](https://github.com/Anita0209/imagenes/blob/main/esfera.png?raw=true)

$$T = T_{1} cuando r = a$$
$$T = T_{2} cuando r = b$$

$$\frac{d}{dr}(r^{2}\frac{dT}{\delta r})=0$$
Integrando obtenemos
$$r^{2}\frac{dT}{dr} = C_{1}$$


$$\frac{dT}{dr} = \frac{C_{1}}{r^{2}}$$

Integrand de nuevo obtenemos

$$T = -\frac{C_{1}}{r}+C_{2} $$

Usando la condición de frontera en r = a, b y b>a se obtiene 
 $$T_{1} = -\frac{C_{1}}{a}+C_{2}$$
 $$T_{2} = -\frac{C_{1}}{b}+C_{2}$$
 
 
 En la siguiente celda esta el código python para estas ecuaciones, ejecutalo y podrás ver las gráficas

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

output_notebook()

# Definir los coeficientes c1 y c2
c1 = 2
c2 = 5

# Definir los valores de a y b puedes cambiar los balores de a y b siempre y cuando b>a
a = 1
b = 2

# Generar algunos valores de x
x = np.linspace(-10, 100)

# Calcular los valores correspondientes de T1 y T2 usando las ecuaciones
T1 = -c1/a + c2
T2 = -c1/b + c2

# Crear una figura
fig = figure(title='Gráfico de dos líneas', x_axis_label='x', y_axis_label='T')

# Agregar una línea para la ecuación T1
fig.line(x, np.full_like(x, T1), legend_label='T1', line_width=2)

# Agregar una línea para la ecuación T2
fig.line(x, np.full_like(x, T2), legend_label='T2', line_color='red', line_width=2)

# Mostrar la figura
show(fig)
```
